```python
class SimpleMemoryAgent:
    def __init__(self):
        self.memory = []

    def fetch_content(self, topic):
        user_data = {
            "ai": "Artificial intelligence helps machines learn, think, and make decisions.",
            "python": "Python is an easy programming language used in AI, web development, and data science.",
            "machine learning": "Machine learning is a part of AI where systems learn from data."
        }

        content = user_data.get(topic.lower())

        if content is None:
            return None

        return {
            "topic": topic,
            "content": content
        }

    def summarize_content(self, text):
        words = text.split()
        short_text = " ".join(words[:8])

        if len(words) > 8:
            short_text += "..."

        return short_text

    def save_memory(self, action, message):
        self.memory.append({
            "action": action,
            "message": message
        })

    def show_memory(self):
        if not self.memory:
            return "Memory is empty."

        lines = []
        for index, item in enumerate(self.memory, start=1):
            lines.append(f"{index}. {item['action']}: {item['message']}")
        return "\n".join(lines)

    def handle_fetch(self, user_input):
        parts = user_input.split(maxsplit=1)

        if len(parts) < 2:
            return "Please type like this: fetch ai"

        topic = parts[1]
        result = self.fetch_content(topic)

        if result is None:
            return f"Sorry, I could not find any data for '{topic}'."

        self.save_memory("fetch", f"Fetched topic '{result['topic']}'")
        self.save_memory("content", result["content"])

        return f"Topic: {result['topic']}\nContent: {result['content']}"

    def handle_summarize(self):
        for item in reversed(self.memory):
            if item["action"] == "content":
                summary = self.summarize_content(item["message"])
                self.save_memory("summary", summary)
                return f"Summary: {summary}"

        return "There is no content in memory to summarize."

    def run(self, user_input):
        text = user_input.lower().strip()

        if text.startswith("fetch"):
            return self.handle_fetch(user_input)

        if text == "summarize":
            return self.handle_summarize()

        if text == "show memory":
            return self.show_memory()

        if text == "exit":
            return "Thank you!"

        return "Unknown command. Try: fetch ai/py,ml, summarize, show memory, or exit."


# Create agent
agent = SimpleMemoryAgent()

# Keep asking user for input
while True:
    user_input = input("Enter command: ")
    output = agent.run(user_input)
    print(output)
    print()

    if user_input.lower().strip() == "exit":
        break
```
