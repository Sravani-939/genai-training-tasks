```python
from sentence_transformers import SentenceTransformer
import faiss

# Passages
passages = [
    "Artificial intelligence is the simulation of human intelligence in machines.",
    "Machine learning is a subset of AI that allows systems to learn from data.",
    "Deep learning uses neural networks with many layers to model complex patterns.",
    "Natural language processing helps computers understand and generate human language.",
    "Embeddings convert text into dense numerical vectors that capture semantic meaning.",
    "FAISS is a library for efficient similarity search over dense vector embeddings.",
    "Chroma is a vector database used to store embeddings and retrieve similar documents.",
    "Sentence Transformers provides pretrained models for creating sentence embeddings."
]

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Generate embeddings
embeddings = model.encode(passages)

# Create FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# Query
query = "Which tool is used for similarity search?"
query_embedding = model.encode([query])

# Search ( result)
_, indices = index.search(query_embedding, 1)

# Print result
print("Query:", query)
print("\nMost similar passage:")
print(passages[indices[0][0]])
```
